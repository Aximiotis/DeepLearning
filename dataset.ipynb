```python
from medmnist import BloodMNIST      # Dataset BloodMNIST
from PIL import Image               # Χειρισμός εικόνων
import matplotlib.pyplot as plt     # Οπτικοποίηση

train_dataset = BloodMNIST(split="train", download=True)  # Training set
test_dataset = BloodMNIST(split="test", download=True)    # Test set
val_dataset = BloodMNIST(split="val", download=True)      # Validation set
```

```python
plt.figure(figsize=(9, 9))   # Δημιουργία καμβά
img = 0

for i in range(0, 100):
    img, label = train_dataset[i]   # Εικόνα και ετικέτα

    plt.subplot(10, 10, i+1)         # Πλέγμα 10x10
    plt.imshow(img, interpolation='nearest')  # Εμφάνιση εικόνας
    plt.axis('off')                  # Απόκρυψη αξόνων
    plt.title(str(label), fontsize=8)  # Τίτλος με label

# plt.tight_layout()
# plt.show()

print(img.size)   # Διαστάσεις εικόνας
print(img.mode)   # Χρωματικό mode (π.χ. RGB)
```

```python
import numpy as np

classes = np.zeros((8, 1))      # Μετρητής κλάσεων για train
test = np.zeros((8, 1))         # Μετρητής κλάσεων για test
validation = np.zeros((8, 1))   # Μετρητής κλάσεων για validation

# Καταμέτρηση labels στο training set
for i in range(0, len(train_dataset)):
    classes[int(train_dataset[i][1])][0] += 1

# Καταμέτρηση labels στο test set
for i in range(0, len(test_dataset)):
    test[int(test_dataset[i][1])][0] += 1

# Καταμέτρηση labels στο validation set
for i in range(0, len(val_dataset)):
    validation[int(val_dataset[i][1])][0] += 1
```

```python
print(classes)
```

```python
ar = np.arange(0, 8)                 # Δείκτες κλάσεων

plt.bar(ar, classes[:, 0])           # Bar plot για training set
plt.xlabel('Κλάση')
plt.ylabel('Πλήθος δεδομένων')
plt.show()

plt.bar(ar, test[:, 0])              # Bar plot για test set
plt.xlabel('Κλάση')
plt.ylabel('Πλήθος δεδομένων')
plt.show()

plt.bar(ar, validation[:, 0])        # Bar plot για validation set
plt.xlabel('Κλάση')
plt.ylabel('Πλήθος δεδομένων')
plt.show()
```
